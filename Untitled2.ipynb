```python
import os

img_path = '/content/dataset_final/images'
lbl_path = '/content/dataset_final/labels'

imgs = {os.path.splitext(f)[0] for f in os.listdir(img_path)}
lbls = {os.path.splitext(f)[0] for f in os.listdir(lbl_path)}

# Wo images jin ka label nahi hai
missing_labels = imgs - lbls

for m in missing_labels:
    for ext in ['.jpg', '.png', '.jpeg']:
        file_to_rem = os.path.join(img_path, m + ext)
        if os.path.exists(file_to_rem):
            os.remove(file_to_rem)

print(f"âœ… Clean up done! Ab images aur labels barabar (320) ho gaye hain.")
```

```python
from ultralytics import YOLO
import yaml

# 1. Dataset Configuration (YAML)
data_config = {
    'path': '/content/dataset_final',
    'train': 'images',
    'val': 'images',
    'nc': 7,
    'names': [
        'Oxygen Tank', 'Nitrogen Tank', 'First Aid Box',
        'Fire Alarm', 'Safety Switch Panel', 'Emergency Phone', 'Fire Extinguisher'
    ]
}

with open('/content/dataset.yaml', 'w') as f:
    yaml.dump(data_config, f)

# 2. YOLOv8 Nano Load Karein (Requirement: < 50ms inference)
model = YOLO('yolov8n.pt')

# 3. Training shuru karein
# Hum 100 epochs kar rahe hain kyunke data kam hai, taake accuracy achi nikal aaye
model.train(
    data='/content/dataset.yaml',
    epochs=100,
    imgsz=416,
    batch=8,
    name='space_safety_final_model'
)
```

```python
from ultralytics import YOLO
import random
import os
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

# 1. Load the best model that was just trained
model = YOLO('/content/runs/detect/space_safety_final_model/weights/best.pt')

# 2. Randomly 3-4 images select karein test karne ke liye
img_list = os.listdir('/content/dataset_final/images')
random_imgs = random.sample(img_list, 3)

print("ðŸ” Testing Model on Random Samples...")

for img_name in random_imgs:
    img_path = os.path.join('/content/dataset_final/images', img_name)

    # Run Prediction
    results = model.predict(source=img_path, conf=0.4) # 40% confidence threshold

    # Visualize the results
    res_plotted = results[0].plot()

    plt.figure(figsize=(10, 6))
    plt.imshow(cv2.cvtColor(res_plotted, cv2.COLOR_BGR2RGB))
    plt.title(f"Detection: {img_name}")
    plt.axis('off')
    plt.show()
```
